```python
df = pd.DataFrame(adata.obsm.X_umap, index = adata.obs.index, 
                  columns = ['UMAP_x','UMAP_y'])#.to_csv('C:/Users/seowjjw/Dropbox/Liver_Paper/Garnett/Umapcoord.csv')
df[['louvain','NormalvsTumor','patientno']] = adata.obs[['louvain','NormalvsTumor','patientno']]
display(df.head())

df.to_csv('./data/Liv73k_phenodata.csv')
```

```python
#adata.raw.var['gene_short_name'] = adata.raw.var.index.tolist()
df = pd.DataFrame(adata.raw.var.index.tolist(), index = adata.raw.var.index.tolist(), columns = ['gene_short_name'])
display(df.head())
```

```python
import mygene 
mg = mygene.MyGeneInfo()

geneSyms = mg.querymany(df.gene_short_name , scopes='symbol', fields='ensembl.gene', species='human')

geneSyms = pd.DataFrame(geneSyms)
geneSyms = geneSyms[['query','ensembl']]
geneSyms = geneSyms[geneSyms.ensembl.notna()]
ensid = []
for x in geneSyms.ensembl:
    if type(x) is dict:
        ensid.extend(list(x.values()))
    elif type(x) is list:
        ensid.extend(list(x[1].values()))
        
geneSyms['ensid'] = ensid
geneSyms = geneSyms[~geneSyms['query'].duplicated()]

geneSyms.head()
```

```python
geneSyms.index = geneSyms['query']
df['ensemblid'] = geneSyms['ensid']
df  = df.loc[~df.ensemblid.isna(),]
display(df.head())
display(df.shape)

df.to_csv('./data/Liv73k_featuredata.csv')
```
